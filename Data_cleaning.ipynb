```python
# Corrected path with raw string
crime = pd.read_csv(r"C:\Users\Digital Suppliers\Downloads\Crime_Data.csv")

# Display the DataFrame
display(crime)
```

```python
crime.info()
```

```python
crime["Date"]= pd.to_datetime(crime["Date"])
```

```python
crime["Updated On"]= pd.to_datetime(crime["Updated On"])
```

```python
crime.describe()
```

```python
crime.isnull().sum()
```

```python
crime.drop('Location',axis=1,inplace=True)
crime.isnull().sum()
```

```python
crime.drop('Ward',axis=1,inplace=True)
crime.drop('Community Area',axis=1,inplace=True)
crime.drop('X Coordinate',axis=1,inplace=True)
crime.drop('Y Coordinate',axis=1,inplace=True)
crime.isnull().sum()
```

```python
crime["Location Description"].fillna("Unknown",inplace=True)
```

```python
crime["Latitude"].fillna(0.0,inplace=True)
crime["Longitude"].fillna(0.0,inplace=True)
```

```python
crime.isnull().sum()
crime.info()
```

```python
crime.to_csv('crime_cleaned.csv',index=False)
```
